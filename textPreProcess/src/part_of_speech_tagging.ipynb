## So basically, if i want to do part of speech tagging. I will do the following.

1. import spacy

2. load en_core_web_sm (spaCy trained pipeline).

3. Get the text, that you'd like to apply POS-tagging on.

4. Give the text to the trained pipeline as an argument.

5. Now get the part of speech and tagging using `.pos_` and `.tag_`

```python
import spacy

nlp = spacy.load("en_core_web_sm")
text = "An apple a day keeps the doctor away."

doc = nlp(text)

for token in doc:
    print(f"{token} <- {token.pos_}")
```

## Named Entity Recognition (NER)

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Alex went to Washington in 1997 and became a Clerk."

doc = nlp(text)

for token in doc:
    print(f"{token} -> {token.ent_type_}")
```

In aboves result many words don't have any result. It is because they are not Names and Entity. so, now in this step we will only print those with NER.

```python
for token in doc:
    if token.ent_type_ != '':
        print(f"{token} -> {token.ent_type_}") 
```

```python
text = "Nepal is having an IT boom and many IT companies like FuseMachine, LeapFrog, DigitalPathshala, etc are coming forward."
doc2 = nlp(text)

for token in doc2:
    print(f"{token} -> {token.ent_type_}") 
```

Problems in above result.
1. Repeated tokens. Tokens shouldn't repeat. Like repeated 'I.T.', ',', etc.
2. Nepal is GPE not PERSON.
3. I.T. is not GPE.

```python
import spacy
from spacy.pipeline import EntityRuler

nlp = spacy.load("en_core_web_sm")
ruler = nlp.add_pipe("entity_ruler",before="ner")

patterns = [
    # {
    #     "label": "GPE",
    #     "pattern": "Nepal",
    # },
    # {
    #     "label": "TECH",
    #     "pattern": "IT"
    # }
    {
        "label": "GPE",
        "pattern": [{"LOWER": "nepal"}]
    },
    {
        "label": "TECH",
        "pattern": [{"LOWER": "it"}]
    }

]

ruler.add_patterns(patterns)

text = "NePaL is having an IT boom and many IT companies like FuseMachine, LeapFrog, DigitalPathshala, etc are coming forward."
doc2 = nlp(text)

for token in doc2:
    # if not(token.is_punct) and token.ent_type_ != '':
     if token.ent_type_ and not token.is_punct:
        print(f"{token} -> {token.ent_type_}") 
```
